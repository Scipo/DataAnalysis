```python
df['priceRangeMin'].fillna(df['priceRangeMin'].mode()[0], inplace=True)
df['priceRangeMax'].fillna(df['priceRangeMax'].mode()[0], inplace=True)
df['menus.amountMax'].fillna(df['menus.amountMax'].mode()[0], inplace=True)
df['menus.amountMin'].fillna(df['menus.amountMin'].mode()[0], inplace=True)
df['latitude'].fillna(df['latitude'].mode()[0], inplace=True)
df['longitude'].fillna(df['longitude'].mode()[0], inplace=True)
```

```python
dt_count_N = df['city'].value_counts()
dt_count_N
```

```python
df_count_menue = df['menus.name'].value_counts()
df_count_menue
```

```python
df_places = df[['city', 'cuisines','menus.name', 'priceRangeMin', 'priceRangeMax']].groupby(['city', 'menus.name']).sum().reset_index()
df_places.sort_values(by='city', ascending=False, inplace=True)
```

```python
fig = px.sunburst(data_frame=df_places, path=['city', 'menus.name',], values='priceRangeMax', title='Menu offers per city')
fig.show()
```

```python
fig = px.bar(df_places, x = 'city', y='priceRangeMax', color='menus.name', title='Prise scale')
fig.show()
```

```python
df_places = df[['city', 'cuisines','name','menus.name', 'priceRangeMax']].groupby(['city', 'cuisines','name']).sum().reset_index()
df_places.sort_values(by='city', ascending=False, inplace=True)
```

```python
fig = px.sunburst(data_frame=df_places, path=['city', 'cuisines','name'], values='priceRangeMax', title='Restaurans per city')
fig.show()
```

```python
top_cities = df.groupby('cuisines').size().reset_index().rename(columns={0: 'Total'}).sort_values('Total', ascending=False).head()
fig = px.pie(top_cities, values='Total', names='cuisines', color_discrete_sequence=px.colors.sequential.RdBu, title='Top 5 cuisines')
fig.show()
```

```python
fig = px.scatter_mapbox(df,
# Here, plotly gets, (x,y) coordinates
lat="latitude",
lon="longitude",
text='name',

                #Here, plotly detects color of series
                #size="postalCode",
                color = "city",
                labels="name",
                

                zoom=12.0,
                center={"lat":40.728032, "lon":-73.98786},
                height=600,
                width=800)
fig.update_layout(mapbox_style='stamen-terrain')
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.update_layout(title_text="Mexican Food Restaurants")
fig.show()
```
